```python
def plot_U_bonds(line_in_csv):

    with open(parameter_dir+"bonds.csv") as file:
        lines = []
        for line in file:
            lines.append(line.split(","))
    
        print(lines[0])
        print(lines[line_in_csv])
        b0 = float(lines[line_in_csv][2])
        kb = float(lines[line_in_csv][3])
        atoms = lines[line_in_csv][0].strip()+"-"+lines[line_in_csv][1].strip()
        
    b=np.linspace(0,0.5) # in nm
    
    U_bonds = 1/2*kb*(b-b0)**2

    fig = plt.figure()
    plt.plot(b,U_bonds)
    plt.title("$U_{bonds}(b) = 0.5 \cdot k_b \cdot(b-b_0)^2$\nwith $k_b$=%.1f kJ mol$^{-1}$nm$^{-2}$ and $b_0$=%.4f nm"%(kb,b0))
    plt.xlabel("$b_{%s}$ / nm"%atoms)
    plt.ylabel("$U_{bonds}$ / kJ mol$^{-1}$")
    #plt.close()

    plt.xlim([0,0.5])
    plt.ylim([0,1e4])

    
    plt.savefig("U_bonds.png", dpi=300)
    plt.close()
    
    return 

def plot_U_angles(line_in_csv):
    
    with open(parameter_dir+"angles.csv") as file:
        lines = []
        for line in file:
            lines.append(line.split(","))
    
        print(lines[0])
        print(lines[line_in_csv])
        th0 = float(lines[line_in_csv][3])
        cth = float(lines[line_in_csv][4])
        atoms = lines[line_in_csv][0].strip()+"-"+lines[line_in_csv][1].strip()+"-"+lines[line_in_csv][2].strip()
        
    th=np.linspace(0,360,361) # in deg
    
    U_angles = 1/2*cth*(th-th0)**2
    
    fig = plt.figure()
    plt.plot(th,U_angles)
    plt.title("$U_{angles}(Î¸) = 0.5 \cdot k_{Î¸} \cdot(Î¸-Î¸_0)^2$\nwith $k_{Î¸}$=%.1f kJ mol$^{-1}$deg$^{-2}$ and $Î¸_0$=%.4f deg"%(cth,th0))
    plt.xlabel("$Î¸_{%s}$ / deg"%atoms)
    plt.ylabel("$U_{angles}$ / kJ mol$^{-1}$")

    plt.xlim([0,360])
    plt.ylim([0,2e7])

    plt.savefig("U_angles.png", dpi=300)
    plt.close()

    return


def plot_U_dihedral(line_in_csv):
    with open(parameter_dir+"dihedrals.csv") as file:
        lines = []
        for line in file:
            lines.append(line.split(","))
    
        print(lines[0])
        print(lines[line_in_csv])
        div = float(lines[line_in_csv][4])

        barrier_term = float(lines[line_in_csv][6])
        phase = float(lines[line_in_csv][5])
        periodicity = float(lines[line_in_csv][7])
        atoms = lines[line_in_csv][0].strip()+"-"+lines[line_in_csv][1].strip()+"-"+lines[line_in_csv][2].strip()+"-"+lines[line_in_csv][3].strip()
    
    
    phi=np.linspace(0,360) # in deg
    
    U_dihedral = barrier_term/div*(1+np.cos(periodicity*phi-phase))
    
    
    plt.figure()
    plt.plot(phi,U_dihedral)
    plt.title("$U_{dihedral}(Ï†) = barrierterm/div \cdot[1+cos(nÏ†-Î´)]$\nwith $barrierterm$=%.3f kJ mol$^{-1}$;div=%i $n=%.3f$; $Î´$=%.4f deg"%(barrier_term,div,periodicity,phase))
    plt.xlabel("$Ï†_{%s}$ / deg"%atoms)
    plt.ylabel("$U_{dihedral}$ / kJ mol$^{-1}$")
    
    plt.xlim([0,360])
    #plt.ylim([0,2e7])
    #plt.close()

    plt.savefig("U_dihedrals.png", dpi=300)
    plt.close()

    return
plot_U_dihedral(6)
```
